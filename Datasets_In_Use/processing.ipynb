```python
import pandas as pd

# loading the csvs files
gdp = pd.read_csv('gdp.csv')
pop = pd.read_csv('population_t.csv')
le = pd.read_csv('life_expectancy_t.csv')

dictionary = {}

# looping the years to generate a nested dictionary
for i in range(1960, 2020):
    dictionary[(str(i), 'gdp')] = gdp[str(i)]
    dictionary[(str(i), 'population')] = pop[str(i)]
    dictionary[(str(i), 'life expectancy')] = le[str(i)]

# generating the dataframe from the dictionary
df = pd.DataFrame(dictionary)
df.index = gdp['Country Name']
df['Total'] = df.sum(axis=1)
df.head()
```

```python
df.to_csv('word_population.csv')
```

```python
gdp_un = pd.read_csv('un_gdp.csv')

d = {
    'Year': [],
    'Country Name': [],
    'Region': [],
    'Type': [],
    'Value': []
}

series = [
    {
        'ds': gdp,
        'serie': 'gdp per capita',
        'country': 'Country Name'
    },
    {
        'ds': pop,
        'serie': 'population',
        'country': 'Entity'
    },
    {
        'ds': le,
        'serie': 'life expectancy',
        'country': 'Entity'
    },
    {
        'ds': gdp,
        'serie': 'gdp',
        'country': 'Country Name'
    }
]

def fill_gdp(country, year, population):
    gdp = gdp_un[gdp_un['Country or Area'] == country]
    gdp = gdp[gdp_un['Year'] == year]

    try:
        gdp_per_capita = gdp.iloc[0]['Value']
        gdp_total = gdp_per_capita * population.iloc[0][str(year)]
        
        return gdp_per_capita, gdp_total
    except:
        return '', ''

def get_country(name):
    for country in countries:
        if country['name'] == name:
          return country['name'], country['region']
    
    return None, None

def process_ds(dataset, series, column_country, year):
    for index, row in dataset.iterrows():
        country, region = get_country(row[column_country])
        if country is not None:
            d['Country Name'].append(country)
            d['Region'].append(region)
            d['Year'].append(year)
            d['Type'].append(series)
        
            if series in ['gdp per capita', 'gdp']:
                population = pop[pop['Entity'] == country]
                if country == 'China':
                    country = 'China (mainland)'
                
                if country in ['Russia', 'Russian Federation']:
                    country = 'Russian Federation'

                if country in ['Iran, Islamic Republic of', 'Iran']:
                    country = 'Iran'

                if country in ['Bolivia (Plurinational State of)', 'Bolivia']:
                    country = 'Bolivia'
                
                if country in ['Viet Nam', 'Vietnam']:
                    country = 'Vietnam'
                
                if country in ["Democratic People's Republic of Korea", 'South Korea']:
                    country = 'Sputh Korea'

                gdp_per_capita, gdp_total = fill_gdp(country, int(year), population)

                if series == 'gdp':
                    d['Value'].append(gdp_total)
                else:
                    d['Value'].append(gdp_per_capita)
            else:
                d['Value'].append(row[str(year)])

for i in range(1970, 2020):
    for ds in series:
        process_ds(ds['ds'], ds['serie'], ds['country'], i)

df = pd.DataFrame(d)
df.to_csv('world_population_series.csv')
```

```python
world_pop = pd.read_csv('world_population_series.csv')

d = {
    'Year': [],
    'gdp': [],
    'gdp per capita': [],
    'population': [],
    'life expectancy': []
}

for i in range(1970, 2020):
    d['Year'].append(i)
    gdp = world_pop[(world_pop["Year"] == i) & (world_pop['Type'] == 'gdp')]
    life_expectancy = world_pop[(world_pop["Year"] == i) & (world_pop['Type'] == 'life expectancy')]
    gdp_per_capita = world_pop[(world_pop["Year"] == i) & (world_pop['Type'] == 'gdp per capita')]
    population = world_pop[(world_pop["Year"] == i) & (world_pop['Type'] == 'population')]
    d['population'].append(population["Value"].sum())
    d['gdp per capita'].append(gdp_per_capita["Value"].sum())
    d['life expectancy'].append(life_expectancy["Value"].sum() / life_expectancy["Value"].count())
    d['gdp'].append(gdp["Value"].sum() / gdp["Value"].count())

dffinal = pd.DataFrame(d)
dffinal.to_csv('world_population_series_forecast.csv')
```
